### Chapter 2 Vector Spaces

Mainly about when \# of Equation < \# of Unknowns

Space: x and y should be closed under addition and scalar multiplication<br>
$ x, y \in \mathbf{V} , c_{1}x + c_{2}y \in \mathbf{V} $<br>
There is a unique "Zero Vector" such that x + 0 = x  . $\forall$ x (원점을 포함하지 않으면 벡터 공간이 될 수 없다.)<br>

Subspace: non empty subset that satisfies the requirements for a vector space.<br>
The origin is smallest subspace.

- The Column space of A<br>
Column space: Set of all Linear combination of columns of Matrix A. $\mathbf{C}(A)$ <br>
$ Ax = b \ \to \ 
\begin{bmatrix}
 | & | &  & | \\
a_{1} &  a_{2} & \cdots & a_{n} \\
 | & | &  & |
\end{bmatrix}
\begin{bmatrix}
x_{1} \\ x_{2} \\ \vdots \\ x_{n}
\end{bmatrix}
= 
\begin{bmatrix}
b_{1} \\ b_{2} \\ \vdots \\ b_{n}
\end{bmatrix}
$<br>
$\to x_{1} \left[ a_{1} \right] + x_{2} \left[ a_{2} \right] + ... + x_{n} \left[ an \right] = \left[ b \right] $ <br>
Ax=b is solvable if b can be expressed as a combination of the columns of A. (b i in column space.) Those Vector b in the spanned space of column vectors of A?

- Nullspace of A<br>
Nullspace of Matrix A consists of all vectors x such that Ax=0 $\mathbf{N}(A)$

#### Solving Ax=0 and Ax=b

Any vector $x_{n}$ in the nullspace can be added to a particular solution $x_{p}$. The solutions to all linear equations have this form, $ x = x_{p} + x_{n} $<br>
Complete solution $ Ax_{p} = b $ and $ Ax_{n} = 0 $ produce  $ A(x_{p}+x_{n}) = b $ 

When the column space doesn't contain every b in $R^{m}$, we need the conditions on b that make $Ax = b$ solvable.
