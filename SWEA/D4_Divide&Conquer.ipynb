    def partition(A,l,r): # 호어파티션
        p = A[l]
        i = l + 1
        j = r
        while i <= j:
            while(i<=j and A[i] <= p): i+=1
            while(i<=j and A[j] >= p): j-=1
            if i <= j:
                A[i], A[j] = A[j], A[i]
        A[l], A[j] = A[j], A[l]
