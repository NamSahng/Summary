    # curValue: 현재까지 담은 물건의 총 가치,  maxValue: 최대 가치
    # n: 물건의 개수

        global maxValue
        if W >= 0:
            if k>n:
                if maxValue < curValue:
                    maxValue = curValue
            else:
                knapsack(W - weight[k], k+1, curValue + value[k])

- DP 적용

a. 부분 문제 정의<br>&nbsp;&nbsp; 
2개의 부분 문제읯 최적해를 사용해서 구함

    K[n][W]: 배낭 무게 W를 1부터 n까지 물건으로 채우는 최대 가치.
    k[n][W-wn]: 배낭 무게 W-wn를 1부터 n-1까지 물건으로 채우는 최대 가치. (case1)
    k[n-1][W]: 배낭 무게 W를 1부터 n-1까지 물건으로 채우는 최대 가치. (case2)
    최적해 = max{case1의 최대 가치 + vn, case2의 최대 가치}
    K[n][W] = max{K[n-1][W-wn] + vn, K[n-1][W]}
    
점화식 표현 <br>
w = 배낭의용량<br>
n개의 물건집합 ={1,2....n}<br>
(vi,wi) = 물건 가치, 무게<br>
K[i][w] = 물건 1~i 까지만 고려하고, (임시) 배낭의 용량이 w일때 최대 가치<br>
    
    <재귀적 정리>
    k[i][w] = 0                             if i = 0 or w =0
            = k[i-1][w]                     if wi > w
            = max(k[i-1][w-wi] + k[i-1][w]) if wi <= w
                   (case1)     (case2)


    <메모이제이션>
    # K[i][W]: 부분문제의 해(최대가치)를 저장하기 위한 리스트, -1로 초기화
    # i: 배낭에 넣을 물건을 나타내는 값(1...W), W: 배낭의 무게
    # n: 물건의 개수

    def knapsack(i,W):
        if K[i][W] != -1:
            return K[i][W]
        if i == 0 or W == 0:
            K[i][W] = 0
            return K[i][W]
        else:
            case1 = 0
            if W >= weight[i]:
                case1 = knapsack(i-1, W - weight[i]) + value[i]

            case2 = knapsack(i-1,W)
            K[i][W] = max(case1,case2)
        return K[i][W]


    <배낭 문제에 DP를 적용해 상향식으로 최적해>
    # 배낭무게 W
    def knapsack():
        for i in range(n+1):
            K[i][0] = 0
        for w in range(W+1):
            K[0][w] = 0

        for i in range(1,n+1):
            for w in range(1, W+1):
                if weight[i] > w:
                    K[i][w] = K[i-1][w]
                else:
                    K[i][w] = max(K[i-1][w - weight[i]], K[i-1][w])
        return K[n][W]
