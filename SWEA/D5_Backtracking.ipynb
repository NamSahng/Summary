    # a[0...n-1]: 집합에 대한 비트포현 저장, 크기는 원소의 수
    # k : 선택한 횟수(현재 노드의 높이), n : 모든 선택수(트리의 높이)

    def subset(a,k,n):
        if k == n:
            process_solution(a,n)
        else:
            a[k] = 0
            subset(a,k+1,n)
            a[k] = 1
            subset(a,k+1,n)

### Permutation

    def permutation(order,k,n):
        if k == n:
            print_order_array(order,n):
        else:
            check = [false] * n
            for i in range(k):
                check[order[i]] = True

            for i in range(n):
                if check[i] == False:
                    order[k] = i
                    permutation(order, k+1, n)
