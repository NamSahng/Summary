### 배낭

* 방법 1: 값이 비싼 물건부터, 최적해 보장 X

* 방법 2: 가장 가벼운 물건, 최적해 보장 X

* 방법 3: 무게당 값이 높은 순서로 , 최적해 보장 X

** 배낭 문제는 보장이 안됨. **

물건을 잘라서 자를 수 있다면: Fractional Knapsack 무게당 가치로 탐욕으로 할 수 있다. 배낭 문제의 상태 공간 트리를 탐색할 때 분기 한정을 위한 한정 값을 구하는 데 사용

### 대표적 탐욕

* 프림

N개의 정접으로 구성된 최소신장트리르 찾음 / 정점을 하나씩 선택하는 과정에서 트리를 확장하면서 MST 찾음 / 그래프


* 크루스칼

N개의 정접으로 구성된 최소신장트리르 찾음 / 싸이클이 없는 서브 그래프들을 확장하면서 MST를 찾음 / 그래프


* 다익스트라

주어진 정접에서 다른 정점들에 대한 최단 경로를 찾음 / 주어진 정점에서 가장 가까운 정점을 선택하면서 출발범에서 다른 모든 정점들의 최단 경로를 찾음 / 그래프

* 허프만 코딩

문서의 압축을 위해 문자들의 빈도수에 따라 코드 값을 부여함 / 출현 빈도가 낮은 문자부터 선택해서 이진 트리를 완성하고 코드값을 부여함 / 문자열
