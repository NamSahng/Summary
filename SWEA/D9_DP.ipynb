### memoization

memoize

1. 작은 부분 문제들의 해들을 구함
2. 해를 이용해 큰 크기의 부분문제 해결
3. 최종 원래 주어진 문제 해결

이는 완전검색을 더 효율적으로 하는방법<br>
Recursive + memoization<br>
점화식을 찾으면 가능(문제를 분석해 재귀적으로 정의하고 수식으로 표현)

동적계획법의 적용요건:<br>
1. 중복 부분문제 구조(Overlapping subproblems)<br>
점화식, 테이블 저장

2. 최적 부분문제 구조(Optimal substructure)<br>
최적화의 원칙을 만족해야만 효율적으로 적용가능<br>
작은문제들도 최적해들로 구성되어야

최적이 안되는 게 ex) 최장경로

분할정복: <br>
큰 문제를 작은 부분문제로 분할, 분할문제를 재귀로 해결, 부문문제 해결합, ex) 병합, 퀵정렬 (보통 하향식)

DP:<br>
부분문제의 해는 더 작은 부분문제의 해를 공유, 모든 부분문제를 한번만 계산하고 결과 저장 후 재사용
** + 동적 계획법에는 부분문제들 사이에 의존적 관계가 존재 **
<br> 함축적인 순서라고도 한다. (보통 상향식)

DP를 적용하기위한방법
1. 최적해 구조 특성파악: 문제를 부분문제로 나눔
2. 최적해의 값을 재귀적으로 정의
3. 상향식 방법으로 최적해 계산

### Coin Change

ex) 동전 1,4,6원 에서 최소 동전

그리디: 6, 1, 1 <br>
최적해: 4, 4 <br>

    def bino2(n,k):
        for i in range(n+1): # i 행값
            for j in range(min(i,k)+1): # j 열값
